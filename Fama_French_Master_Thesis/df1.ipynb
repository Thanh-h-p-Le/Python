```python
sorting_data = pd.read_csv('Sorting_2023.csv')
stock_price_data = pd.read_csv('Stock_price_2023.csv')
```

```python
stock_price_data['Dates'] = pd.to_datetime(stock_price_data['Dates'], dayfirst=True)
```

```python
stock_price_data.sort_values('Dates', inplace = True)
stock_price_data.set_index('Dates', inplace=True)
```

```python
annual_returns = stock_price_data.pct_change(periods = 252).dropna(how = 'all').iloc[-1]
sorting_data['Ticker'] = sorting_data['Ticker'].str.replace(' ', '').str.upper()
annual_returns.index = annual_returns.index.str.replace(' ', '').str.upper()

sorting_data['12M_return'] = sorting_data['Ticker'].map(annual_returns)

sorting_data.dropna(subset=['12M_return'], inplace = True)

percentiles = sorting_data['12M_return'].quantile([0.33,0.66])
sorting_data['Momentum'] = pd.cut(sorting_data['12M_return'],
                                  bins = [-float('inf'), percentiles[0.33], percentiles[0.66], float('inf')],
                                  labels=['Low', 'Medium', 'High'])
```

```python
high_momentum = sorting_data[sorting_data['Momentum'] == 'High']
low_momentum = sorting_data[sorting_data['Momentum'] == 'Low']

high_return = (high_momentum['12M_return'] * high_momentum['Market_cap']).sum()/high_momentum['Market_cap'].sum()
low_return = (low_momentum['12M_return'] * low_momentum['Market_cap']).sum() / low_momentum['Market_cap'].sum()
```
