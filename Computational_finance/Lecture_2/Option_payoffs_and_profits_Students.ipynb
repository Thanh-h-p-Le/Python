# Option Payoffs and Profits

We can have four different option positions which have the following payoff and profit profiles:

Long call
- Payoff: $\max(S - E, 0)$
- Profit: $\max(S - E, 0) - C$

Long put
- Payoff: $\max(E - S, 0)$
- Profit: $\max(E - S, 0) - P$

Short call
- Payoff: $- \max(S - E, 0)$
- Profit: $C - \max(S - E, 0)$

Short put
- Payoff: $- \max(E - S, 0)$
- Profit: $P - \max(E - S, 0)$

```python
# Import libraries and packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Payoff and Profit Diagramms

```python
# Inputs

E = 100.00
C = 10.00
P = 5.00
S_T = np.arange(0, 201, 1)
```

### Long Call

```python
# Empty lists for storing the output of the calculations

output_payoff = []
output_profit = []

# Loop through all potential stock prices at experation S_T
for S in S_T:
    payoff = max(S - E, 0)
    output_payoff.append(payoff)
    profit = max(S - E, 0) - C
    output_profit.append(profit)

# Plot results

plt.axhline(y = 0, linewidth = 1, color = 'black', ls = '--')
plt.axvline(x = 100, linewidth = 1, color = 'black', ls = '--')
plt.plot(S_T, output_payoff, label = 'Payoff')
plt.plot(S_T, output_profit, label = 'Profit')
plt.title('Long Call')
plt.xlabel('$S_T$')
plt.ylabel('USD')
plt.legend();
```
