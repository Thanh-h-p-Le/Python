# Coin Tossing Simulation

One of the simplest random processes is tossing a coin: We start with a value of 100 and toss a coin. If we throw head (tail), we multiply the value by 1.01 (0.99) $\rightarrow$ after one toss, the value will be either 99 or 100. We toss again and multiply the previous result again by either 1.01 or 0.99, depending on whether we throw head or tail. We will now observe one of three potential outcomes:

1) 2 heads so that the value is $1.01^2 \times 100$, \
2) 2 tails so that the value is $0.99^2 \times 100$, \
3) 1 head or 1 tail (or 1 tail and 1 head) so that the value is $1.01 \times 0.99 \times 100 = 0.99$ (or equivalently $0.99 \times 1.01 \times 100 = 0.99$.

Python has various modules we can use to simulate a coin toss. For single random draws we can use

1) random.randint(0, 1) which returns a random integer in the interval [0,1], \
2) random.randrange(0, 2) which returns a random integer in the interval [0,2).

As we are interested in a series of draws, we need to store them in an array and can use

1) np.random.randint(0, 2, size = size) which returns an array of random draws from [0,2) of length size, \
2) np.random.choice([0, 1], p = [0.5, 0.5], size = size) which returns an array of random draws from [0, 1] of length size.

```python
# Import libraries and packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## Single Random Draws

```python
random.randint(0, 1)
```

```python
random.randrange(0, 2)
```

## Series of Random Draws

```python
n_flips = 20
flip_results = np.random.randint(0, 2, size = n_flips)
print(flip_results)
```

```python
n_flips = 20
p_head = 0.5
p_tail = 1 - p_head
flip_results = np.random.choice([0, 1], p = [p_head, p_tail], size = n_flips)
print(flip_results)
```

## Function to Plot Coin Tossing Simulation

```python
def Cointoss(n):
    # Prob for head and tail
    p_head = 0.5
    p_tail = 1 - p_head
    
    # Starting values
    x = 0
    y = 100

    # Generating trajectory
    flips = np.arange(n + 1)
    positions = [y]
    directions = ["head", "tail"]

    for i in range(1, n + 1):
        # Randomly select either tail or head
        step = np.random.choice(directions, p = [p_head, p_tail])

        # Move the object up or down
        if step == "head":
            y *= 0.99
        elif step == "tail":
            y *= 1.01
        
        # Keep track of positions
        positions.append(y)

    return flips, positions

flip_data, position_data = Cointoss(1000)
plt.plot(flip_data, position_data)
plt.title('Coin Tossing Simulation')
plt.xlabel('Number of Coin Flips')
plt.ylabel('')
plt.show;
```
