## Payoff and Profit Diagramms

### Protective Put

A protective put consits of long 1 stock and long 1 ATM put. The payoffs and profits are

Long 1 share of stock
- Payoff: $S_T$
- Profit: $S_T - S_t$

Long 1 ATM put
- Payoff: $\max(E - S_T, 0)$
- Profit: $\max(E - S_T, 0) - P$

Protective put
- Payoff: $S_T + \max(E - S_T, 0)$
- Profit: $S_T - S_t + \max(E - S_T, 0) - P$

```python
# Inputs
E = 100
P = 5
S_T = np.arange(0, 201, 1)
```

```python
# Empty lists for storing the output of the calculations

output_payoff_stock = []
output_profit_stock = []

output_payoff_put = []
output_profit_put = []

output_payoff_strategy = []
output_profit_strategy = []

# Loop through all potential stock prices at expiration S_T

for S in S_T:
    payoff_stock = S_T
    output_payoff_stock.append(payoff_stock)
    profit_stock = S - 100
    output_profit_stock.append(profit_stock)

    payoff_put = max(E - S, 0)
    output_payoff_put.append(payoff_put)
    profit_put = max(E - S, 0) - P
    output_profit_put.append(profit_put)

    payoff_strategy = S + max(E - S, 0)
    output_payoff_strategy.append(payoff_strategy)
    profit_strategy = S - 100 + max(E - S, 0) - P
    output_profit_strategy.append(profit_strategy)
```

```python
# Payoff
plt.axhline(y = 0, linewidth = 1, color = 'black', ls = '--')
plt.axvline(x = 100, linewidth = 1, color = 'black', ls = '--')
plt.plot(S_T, output_profit_stock, label = 'Profit Stock')
plt.plot(S_T, output_profit_put, label = 'Profit Put')
plt.plot(S_T, output_profit_strategy, label = 'Profit Protective Put')
plt.title('Protective Put - Payoff')
plt.xlabel('$S_T$')
plt.ylabel('$USD$')
plt.legend()
```

### Bull Spread

A bull spread consits of 2 call options on the same underlying with the same maturity but two different strikes, where $E_1 < E_2$:
- Long 1 call option with the lower strike $E_1$
- Short 1 call option with the higher strike $E_2$

The payoffs and profits are

Long 1 call option with the lower strike
- Payoff: $\max(S - E_1, 0)$
- Profit: $\max(S - E_1, 0) - C_1$

Short 1 call option with the higher strike
- Payoff: $- \max(S - E_2, 0)$
- Profit: $C_2 - \max(S - E_2, 0)$

Bull spread
- Payoff: $\max(S - E_1, 0) - \max(S - E_2, 0)$
- Profit: $\max(S - E_1, 0) - C_1 + C_2 - \max(S - E_2, 0)$

### Bear Spread

A bear spread consits of 2 put options on the same underlying with the same maturity but two different strikes, where $E_1 < E_2$:
- Short 1 put option with the lower strike $E_1$
- Long 1 put option with the higher strike $E_2$

The payoffs and profits are

Short 1 put option with the lower strike
- Payoff: $- \max(E_1 - S, 0)$

Long 1 put option with the higher strike
- Payoff: $\max(E_2 - S, 0)$
- Profit: $\max(E_2 - S, 0) - P_2$

Bear spread
- Payoff: $- \max(E_1 - S, 0) + \max(E_2 - S, 0)$
- Profit: $P_1 - \max(E_1 - S, 0) + \max(E_2 - S, 0) - P_2$

### Butterfly

A butterfly spread consits of 4 call options on the same underlying with the same maturity but three different strikes, where $E_1 < E_2 < E_3$:
- Long 1 call option with the lowest strike $E_1$
- Short 2 call options with the strike inbetween $E_2$
- Long 1 call option with the highest strike $E_3$

The payoffs and profits are

Long 1 call option with the lowest strike
- Payoff: $\max(S - E_1, 0)$
- Profit: $\max(S - E_1, 0) - C_1$

Short 2 call options with the strike inbetween
- Payoff: $2 \times [- \max(S - E_2, 0)]$
- Profit: $2 \times [C_2 - \max(S - E_2, 0)]$

Long 1 call option with the highest strike
- Payoff: $\max(S - E_3, 0)$
- Profit: $\max(S - E_3, 0) - C_3$

Butterfly
- Payoff: $\max(S - E_1, 0) + 2 \times [- \max(S - E_2, 0)] + \max(S - E_3, 0)$
- Profit: $\max(S - E_1, 0) - C_1 + 2 \times [C_2 - \max(S - E_2, 0)] + \max(S - E_3, 0) - C_3$
