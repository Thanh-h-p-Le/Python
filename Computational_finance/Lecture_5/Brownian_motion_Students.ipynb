## Symmetric Random Walk

Consider a coin-tossing experiment where we have successive coin tosses $\omega = \omega_1, \omega_2, ..., \omega_n$, where $\omega_n$ denotes the outcome of the $n$-th toss:
- $R_j = 
\begin{cases}
     1   & \text{if } \omega_j = H \\
    -1   & \text{if } \omega_j = T \\
\end{cases}$
- If $S_0 = 0$, total amount of money we have won up to and including the $i$-th coin toss is
\begin{equation}
S_i = \sum_{j = 1}^{i} R_j
\end{equation}

- Computed path by path, one increment at a time
- Because we will move either +1 or -1 after each toss, $|S_j - S_{j-1}| = 1$ so that the quadratic variation is always $i$
\begin{equation}
\sum_{j = 1}^i (S_j - S_{j-1})^2 = i

## Brownian Motion

We can make $n$ larger and larger. The limiting process for this random walk as the step size goes to zero is called Brownian motion and denoted by $X(t)$.

As Brownian motion is the continuous time limit of the discrete time random walk. the pathrs are continuous.

Over finite time increments $t_{i-1}$ to $t_i$, $X(t_i) - X(t_{i-1})$ is normally distributed with mean zero and variance $t_i - t_{i-1}$.

The quadratic variation will almost surely converge towards $t$.

## Brownian Motion with Drift

General form of Brownian motion with drift is $dS = \mu dt + \sigma dX$. This stochastic differential equation can be integrated exactly to get $S(t) = S(0) + \mu t + \sigma (X(t) - X(0))$.

## Geometric Brownian Motion

For a positive starting value $S(0) > 0$, the dynamics of the process are given by the following stochastic differential equation:
\begin{equation}
dS = \mu S dt + \sigma S dX
\end{equation}

The explicit expression of the integrated form is
\begin{equation}
S(t) = S(0) e^{(\mu - \frac{1}{2} \sigma^2)t + \sigma (X(t) - X(0))}
\end{equation}
