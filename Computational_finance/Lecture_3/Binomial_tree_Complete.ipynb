```python
# Import libraries and packages

import finoptions as fo
import numpy as np
```

```python
# Inputs

S     = 100
K     = 100
t     = 1
r     = 10 / 100
b     = 0 / 100
sigma = 20 / 100
n     = 12
```

```python
# European options
opt_european = fo.binomial_tree_options.CRRBinomialTreeOption(S = S, K = K, t = t, r = r, b = b, sigma = sigma, n = n, type = 'european')
opt_european = fo.binomial_tree_options.CRRBinomialTreeOption()
## Call
opt_european_call = opt_european.call()

## Put
opt_european_put = opt_european.put()


# Amerian options
opt_american = fo.binomial_tree_options.CRRBinomialTreeOption(S = S, K = K, t = t, r = r, b = b, sigma = sigma, n = n, type = 'american')

## Call
opt_american_call = opt_american.call()

## Put
opt_american_put = opt_american.put()
```

```python
# Access option values

print(opt_european_call)
print(opt_european_put)
print(opt_american_call)
print(opt_american_put)
```

```python
# Access model summaries

opt_european.summary()
opt_american.summary()
```

```python
# Create objects with entire tree of option values

opt_european_call_tree = opt_european.call(tree = True)
opt_european_put_tree = opt_european.put(tree = True)
opt_american_call_tree = opt_american.call(tree = True)
opt_american_put_tree = opt_american.put(tree = True)
```

```python
# Access trees of option values

print(opt_european_call_tree)
```

```python
print(opt_european_put_tree)
```

```python
# Plot trees of option values

opt_european.plot(call = True)
opt_american.plot(call = True);
```

```python
# Calculate Greeks

opt_european_greeks = opt_european.greeks()
opt_american_greeks = opt_american.greeks()
print(opt_european_greeks)
print(opt_american_greeks)
```
