```python
# Import libraries and packages
import finoptions as fo
import numpy as np
```

# 25% out-of-the-money call options with a monthly time interval

```python
# Inputs
S = 147.42 # Stock price
K = 185 # Strike price
t = 1 # Time to expiration: 1 year
r = 0.048 # Risk-free rate
b = 0 # Cost of carry
sigma = 0.3430 # Implied volatility
n = 12 # Number of steps representing monthly time interval

# Contract name: AMZN250117C00185000
```

```python
# European options
opt_european_1 = fo.binomial_tree_options.CRRBinomialTreeOption(S = S, K = K, t = t, r = r, b = b, sigma = sigma, n = n, type = 'european')

## Call
opt_european_call_1 = opt_european_1.call()
```

```python
# Access option values

print(opt_european_call_1)
```

```python
# Access model summaries
opt_european_1.summary()
```

```python
# Create objects with entire tree of option values
opt_european_call_tree_1 = opt_european_1.call(tree = True)
```

```python
# Access trees of option values
print(opt_european_call_tree_1)
```

```python
# Plot trees of option values

opt_european_1.plot(call = True);
```

```python
# Calculate Greeks

opt_european_greeks_1 = opt_european_1.greeks()
opt_european_greeks_1
```
