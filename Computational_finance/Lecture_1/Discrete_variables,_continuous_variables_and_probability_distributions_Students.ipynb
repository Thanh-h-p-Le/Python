# Discrete Variables, Continuous Variables and Probability Distributions

A probability distribution is a function that describes the probability of obtaining the possible values of a variable. Depending on whether a variable has discrete outcomes or continuous outcomes, we will get a discrete distribution represented by the probability mass function (pmf) or a continuous distribution represented by the probability density function (pdf).

```python
# Import libraries and packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns
```

## Probability Density Function of Normally Distributed Random Variable

### PDF for N(0,1) Random Variable

```python
def PlotNormalPDF():
    mu = 0.0
    sigma = 1.0

    x = np.linspace(-5, 5, 100)

    plt.figure()
    plt.plot(x, st.norm.pdf(x, mu, sigma))
    plt.title('Probability Densisty Function N(0, 1) Random Variable')
    plt.xlabel('x')
    plt.ylabel('PDF')

PlotNormalPDF();
```
